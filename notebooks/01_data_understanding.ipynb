```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns

%matplotlib inline 

print("All libraries imported successfully")
```

```python
df = pd.read_csv('../data/raw/hotel_bookings.csv')
print("Dataset loaded successfully")
print(f"Rows: {df.shape[0]}")
print(f"columns: {df.shape[1]}")
```

```python
print("Missing values in each column")
print(df.isnull().sum())
```

```python
df['children'] = df['children'].fillna(0)
df['country'] = df['country'].fillna('unknown')
df['agent'] = df['agent'].fillna(0)
df['company'] = df['company'].fillna(0)

print("missing values fixed!")
print(df.isnull().sum().sum(), "total missing values remaining")
```

```python
print("Any missing values left?", df.isnull().sum().sum())
```

```python
df['total_nights'] = df['stays_in_weekend_nights'] + df['stays_in_week_nights']
df['total_guests'] = df['adults'] + df['children'] + df['babies']

print("New columns created!")
print(df[['total_nights','total_guests']].head())
```

```python
month_map = {
    'January':1, 'February':2, 'March':3, 'April':4, 'May':5, 'June':6, 'July':7, 'August':8, 'September':9, 'October':10,
    'November':11, 'December':12
}
df['arrival_month_num'] = df['arrival_date_month'].map(month_map)

print("Month numbers created!")
print(df[['arrival_date_month','arrival_month_num']].head(12))
```

```python
df.to_csv('../data/processed/hotel_bookings_cleaned.csv', index=False)
print("Cleaned dataset saved successfully!")
print("Location: data/processed/hotel_bookings_cleaned.csv")
```
