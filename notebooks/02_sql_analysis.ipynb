```python
import sqlite3
import pandas as pd
print("libraries imported successfully")
```

```python
df = pd.read_csv('../data/processed/hotel_bookings_cleaned.csv')
conn = sqlite3.connect('../data/processed/hotel.db')
df.to_sql('bookings',conn, if_exists='replace', index=False)
print("Database created successfully")
print("Table 'booking' is ready to query")
```

```python
query1 = """
SELECT
hotel,
count(*) as total_bookings,
SUM(is_canceled) as total_cancellations,
ROUND(AVG(is_canceled)*100,2) as cancellation_rate
FROM bookings
GROUP BY hotel
ORDER BY total_bookings DESC
"""
result1 = pd.read_sql(query1,conn)
print("Query1: Bookings by hotel type")
result1
```

```python
query2= """
SELECT
arrival_date_month,
arrival_month_num,
COUNT(*) as total_bookings,
SUM(is_canceled) as total_cancellations,
ROUND(AVG(is_canceled)*100,2) as cancellation_rate
FROM bookings
GROUP BY arrival_date_month, arrival_month_num
ORDER BY arrival_month_num ASC
"""
result2 = pd.read_sql(query2, conn)
print("Query2: Monthly booking trends")
result2
```

```python
query3= """
SELECT
arrival_date_month,
arrival_month_num,
hotel,
count(*) as total_bookings,
ROUND(AVG(is_canceled)*100,2) as cancellation_rate
FROM bookings
GROUP BY arrival_date_month, arrival_month_num, hotel
ORDER BY arrival_month_num ASC
"""
result3 = pd.read_sql(query3,conn)
print("query3: Monthly Bookings by Hotel Type")
result3
```

```python
query4= """
SELECT
market_segment,
COUNT(*) as total_bookings,
SUM(is_canceled) as total_cancellations,
ROUND(AVG(is_canceled)*100,2) as cancellation_rate
FROM bookings
GROUP BY market_segment
ORDER BY cancellation_rate DESC
"""

result4 = pd.read_sql(query4,conn)
print("Query4: Cancellation by Market segment")
result4
```

```python
query5= """
SELECT
customer_type,
COUNT(*) as total_bookings,
SUM(is_canceled) as total_cancellations,
ROUND(AVG(is_canceled)*100,2) as cancellation_rate
FROM bookings
GROUP BY customer_type
ORDER BY cancellation_rate DESC
"""

result5= pd.read_sql(query5, conn)
print("Query5: Cancellation by Customer Type")
result5
```

```python
query6= """
SELECT
distribution_channel,
COUNT(*) as total_bookings,
SUM(is_canceled) as total_cancellations,
ROUND(AVG(is_canceled)*100,2) as cancellation_rate
FROM bookings
GROUP BY distribution_channel
ORDER BY cancellation_rate DESC
"""

result6= pd.read_sql(query6, conn)
print("Query6: Cancellation by Distribution Channel")
result6
```

```python
query7= """
SELECT
deposit_type,
COUNT(*) as total_bookings,
SUM(is_canceled) as total_cancellations,
ROUND(AVG(is_canceled)*100,2) as cancellation_rate
FROM bookings
GROUP BY deposit_type
ORDER BY cancellation_rate DESC
"""

result7 = pd.read_sql(query7,conn)
print("Query 7:Cancellation by Deposit Type")
result7
```

```python
query8="""
SELECT
CASE
WHEN lead_time <=7 THEN'1. Last minute (0-7 days)'
WHEN lead_time <=30 THEN'2. Short (8-30 days)'
WHEN lead_time <=90 THEN'3. Medium (31-90 days)'
When lead_time <=180 THEN'4. Lonng (91-180 days)'
ELSE                     '5. Very long(180+ days)'
END as lead_time_band,
COUNT(*) as total_bookings,
SUM(is_canceled) as total_cancellations,
ROUND(AVG(is_canceled)*100,2) as cancellation_rate
FROM bookings
GROUP BY lead_time_band
ORDER BY lead_time_band ASC
"""
result8 = pd.read_sql(query8,conn)
print("Query8: Cancellation by Lead Time Band")
result8
```

```python
query9 = """
SELECT 
    total_of_special_requests,
    COUNT(*) as total_bookings,
    SUM(is_canceled) as total_cancellations,
    ROUND(AVG(is_canceled) * 100, 2) as cancellation_rate
FROM bookings
GROUP BY total_of_special_requests
ORDER BY total_of_special_requests ASC
"""

result9 = pd.read_sql(query9, conn)
print("Query 9: Cancellation by Special Requests")
result9
```

```python
query10 = """
SELECT 
    hotel,
    arrival_date_month,
    arrival_month_num,
    ROUND(AVG(adr), 2) as avg_daily_rate,
    COUNT(*) as total_bookings
FROM bookings
WHERE is_canceled = 0
GROUP BY hotel, arrival_date_month, arrival_month_num
ORDER BY hotel ASC, arrival_month_num ASC
"""

result10 = pd.read_sql(query10, conn)
print("Query 10: ADR by Hotel Type and Month")
result10
```

```python

query11 = """
SELECT 
    customer_type,
    ROUND(AVG(adr), 2) as avg_daily_rate,
    COUNT(*) as total_bookings,
    ROUND(AVG(is_canceled) * 100, 2) as cancellation_rate
FROM bookings
GROUP BY customer_type
ORDER BY avg_daily_rate DESC
"""

result11 = pd.read_sql(query11, conn)
print("Query 11: ADR by Customer Type")
result11
```

```python
query12 = """
SELECT 
    CASE
        WHEN total_nights = 0 THEN '0. No Stay (0 nights)'
        WHEN total_nights <= 2 THEN '1. Short (1-2 nights)'
        WHEN total_nights <= 5 THEN '2. Medium (3-5 nights)'
        WHEN total_nights <= 10 THEN '3. Long (6-10 nights)'
        ELSE                        '4. Extended (10+ nights)'
    END as stay_length,
    COUNT(*) as total_bookings,
    ROUND(AVG(adr), 2) as avg_daily_rate,
    ROUND(AVG(is_canceled) * 100, 2) as cancellation_rate
FROM bookings
GROUP BY stay_length
ORDER BY stay_length ASC
"""

result12 = pd.read_sql(query12, conn)
print("Query 12: ADR by Stay Length")
result12
```

```python
query13 = """
SELECT 
    country,
    COUNT(*) as total_bookings,
    SUM(is_canceled) as total_cancellations,
    ROUND(AVG(is_canceled) * 100, 2) as cancellation_rate,
    ROUND(AVG(adr), 2) as avg_daily_rate
FROM bookings
WHERE country != 'Unknown'
GROUP BY country
ORDER BY total_bookings DESC
LIMIT 10
"""

result13 = pd.read_sql(query13, conn)
print("Query 13: Top 10 Countries by Bookings")
result13
```

```python
query14 = """
SELECT 
    CASE
        WHEN previous_cancellations = 0 THEN 'No Previous Cancellations'
        WHEN previous_cancellations = 1 THEN '1 Previous Cancellation'
        WHEN previous_cancellations <= 3 THEN '2-3 Previous Cancellations'
        ELSE '4+ Previous Cancellations'
    END as cancellation_history,
    COUNT(*) as total_bookings,
    SUM(is_canceled) as total_cancellations,
    ROUND(AVG(is_canceled) * 100, 2) as cancellation_rate
FROM bookings
GROUP BY cancellation_history
ORDER BY cancellation_rate ASC
"""

result14 = pd.read_sql(query14, conn)
print("Query 14: Effect of Previous Cancellations")
result14
```

```python
query15 = """
SELECT 
    market_segment,
    COUNT(*) as total_bookings,
    ROUND(AVG(adr), 2) as avg_daily_rate,
    ROUND(AVG(is_canceled) * 100, 2) as cancellation_rate,
    ROUND(AVG(adr) * (1 - AVG(is_canceled)), 2) as true_value_per_booking
FROM bookings
GROUP BY market_segment
ORDER BY true_value_per_booking DESC
"""

result15 = pd.read_sql(query15, conn)
print("Query 15: Most Valuable Market Segments")
result15
```
